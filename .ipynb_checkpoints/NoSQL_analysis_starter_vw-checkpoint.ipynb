## Notebook Set Up

```python
# Import dependencies
from pymongo import MongoClient
import pandas as pd
from pprint import pprint
```

```python
# Create an instance of MongoClient
mongo = MongoClient(port=27017)
```

```python
# Assign the uk_food database to a variable name
uk_food_db = mongo['uk_food']
```

```python
# Review the collections in our database
print(uk_food_db.list_collection_names())
```

```python
# Assign the collection to a variable
establishments = uk_food_db['establishments']
```

## Part 3: Exploratory Analysis

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
hygiene_query = {'scores.Hygiene': {'$eq': 20}}
hygiene_results = list(establishments.find(hygiene_query))
hygiene_results
```

```python
# Use count_documents to display the number of documents in the result
hygiene_count = establishments.count_documents(hygiene_query)
hygiene_count
```

```python
# Display the first document in the results using pprint
pprint(hygiene_results[0])
```

```python
# Convert the result to a Pandas DataFrame
hygiene_df = pd.DataFrame(hygiene_results)
```

```python
# Display the number of rows in the DataFrame
print("Rows in DataFrame: ", len(hygiene_df))
```

```python
# Display the first 10 rows of the DataFrame
hygiene_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
london_query = {'LocalAuthorityName': {'$regex': 'London'},
               'RatingValue': {'$gte': '4'}}
london_results = list(establishments.find(london_query))
london_results
```

```python
# Use count_documents to display the number of documents in the result
london_count = establishments.count_documents(london_query)
london_count
```

```python
# Display the first document in the results using pprint
pprint(london_results[0])
```

```python
# Convert the result to a Pandas DataFrame
london_df = pd.DataFrame(london_results)
```

```python
# Display the number of rows in the DataFrame
print("Rows in DataFrame: ", len(london_df))
```

```python
# Display the first 10 rows of the DataFrame
london_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of '5', sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?
Note: longitude and latitude values were switched in the original starter code

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01000000
latitude = 51.49014200
longitude = 0.08384000

top_query = {'$match': {'geocode.latitude': {'$gte': 51.048014200, '$lte': 51.500142},
                        'geocode.longitude': {'$gte': 0.07384000, '$lte': 0.09384000},
                        'RatingValue': {'$eq': '5'}
                        }
            }
top_sort =  {'$sort': {'scores.Hygiene': 1 }}

# Print the results
top_pipeline = [top_query, top_sort]
top_results = list(establishments.aggregate(top_pipeline))
print(top_results)
```

```python
# Convert result to Pandas DataFrame
top_df = pd.DataFrame(top_results)
top_df.head(5)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest
hygieneLA_query = {'$match': {'scores.Hygiene': {'$eq': 0}}}
hygieneLA_group = {'$group': {'_id': '$LocalAuthorityName', 'count': {'$sum': 1}}}
hygieneLA_sort = {'$sort': {'count': -1}}
hygieneLA_pipeline = [hygieneLA_query, hygieneLA_group, hygieneLA_sort]

# Print the number of documents in the result
hygieneLA_results = list(establishments.aggregate(hygieneLA_pipeline))
print('Number of documents: ', len(hygiene_results))
```

```python
# Print the first 10 results
pprint(hygieneLA_results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
hygieneLA_df = pd.DataFrame(hygieneLA_results)
```

```python
# Display the number of rows in the DataFrame
len(hygiene_df)
```

```python
# Display the first 10 rows of the DataFrame
hygieneLA_df.head(10)
```
